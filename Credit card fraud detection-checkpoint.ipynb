## creditcard.csv

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, BatchNormalization
from keras.callbacks import EarlyStopping
import seaborn as sns
from keras import losses, optimizers, metrics
```

```python
# Load data
df = pd.read_csv('creditcard(1).csv')
```

```python
df = df.drop(['Time'], axis = 1)
rob_scaler = RobustScaler()
df['scaled_amount'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df.drop(['Amount'], axis=1, inplace=True)
scaled_amount = df['scaled_amount']
df.drop(['scaled_amount'], axis=1, inplace=True)
df.insert(0, 'scaled_amount', scaled_amount)
```

```python
# Separate features and target
x = df.drop(['Class'], axis=1).values
y = df['Class'].values

# Separate normal and abnormal events
normal_events = df[df['Class'] == 0].drop(['Class'], axis=1)
abnormal_events = df[df['Class'] == 1].drop(['Class'], axis=1)
```

```python
# Scaling data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(normal_events)

# Train-test split
train_data, test_data = train_test_split(scaled_data, test_size=0.2, random_state=42)
```

```python
# Model definition
n_features = x.shape[1]

encoder = Sequential(name='encoder')
encoder.add(Dense(units=20, input_shape=(n_features,), activation='relu'))
encoder.add(BatchNormalization())
encoder.add(Dropout(0.1))
encoder.add(Dense(units=10, activation='relu'))
encoder.add(BatchNormalization())
encoder.add(Dense(units=5, activation='relu'))

decoder = Sequential(name='decoder')
decoder.add(Dense(units=10, input_shape=(5,), activation='relu'))
decoder.add(BatchNormalization())
decoder.add(Dense(units=20, activation='relu'))
decoder.add(Dropout(0.1))
decoder.add(Dense(units=n_features, activation='sigmoid'))

autoencoder = Model(inputs=encoder.input, outputs=decoder(encoder.output))
autoencoder.compile(loss=losses.mean_squared_error, optimizer=optimizers.Adam(), metrics=[metrics.mean_squared_error])
# Early stopping callback
es = EarlyStopping(monitor='val_loss', min_delta=0.0001, patience=20, restore_best_weights=True)
```

```python
# Training the model
history = autoencoder.fit(x=train_data, y=train_data, epochs=100, verbose=1, validation_data=(test_data, test_data), callbacks=[es], batch_size = 1024)

# Plotting training history
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# Saving the model
autoencoder.save('autoencoder.h5')
```

```python
# Calculate MSE threshold for anomaly detection
train_predicted_x = autoencoder.predict(train_data)
train_events_mse = np.mean(np.square(train_data - train_predicted_x), axis=1)
cut_off = np.percentile(train_events_mse, 95)
print('Threshold (95th percentile MSE):', cut_off)
```

```python
# Plotting MSE for normal and abnormal events
plot_samples = 100
real_x_normal = test_data[:plot_samples]
predicted_x_normal = autoencoder.predict(real_x_normal)
mse_normal = np.mean(np.square(real_x_normal - predicted_x_normal), axis=1)
normal_events_df = pd.DataFrame({'mse': mse_normal, 'n': np.arange(plot_samples), 'anomaly': np.zeros(plot_samples)})

real_x_abnormal = scaler.transform(abnormal_events)[:plot_samples]
predicted_x_abnormal = autoencoder.predict(real_x_abnormal)
mse_abnormal = np.mean(np.square(real_x_abnormal - predicted_x_abnormal), axis=1)
abnormal_events_df = pd.DataFrame({'mse': mse_abnormal, 'n': np.arange(plot_samples), 'anomaly': np.ones(plot_samples)})

mse_df = pd.concat([normal_events_df, abnormal_events_df])

plt.figure(figsize=(10, 8))
sns.lineplot(x='n', y='mse', hue='anomaly', data=mse_df)
plt.axhline(y=cut_off, color='r', linestyle='--', label='Threshold')
plt.title('MSE for Normal and Abnormal Events')
plt.xlabel('Sample Index')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
